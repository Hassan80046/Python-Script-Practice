```python
from IPython.display import clear_output
board = [' ']*10
position=0
play=True
game_on=True


def player_input():
    '''Player 1 will select the marker first i.e. X or O'''
    clear_output()
    marker=''
    print('Welcome to Tic Tac Toe!')
    marker=input('Player 1: Do you want to be X or O: ').upper()
    while marker!='X' and marker!='O':
        print('Please Enter X or O: ')
        marker=input('Player 1: Do you want to be X or O: ')
    player1=marker
    if player1=='X':
        player2='O'
    else:
        player2='X'
    return (player1,player2) #Return Player1,Player2 markers for later use


def choose_first():
    '''Randomly chooses which player will go first'''
    import random
    rand=random.randint(1,2)
    if rand==1:
        return 'Player1'
    elif rand==2:
        return 'Player2'

def player_choice(board):
    '''Player choosing position to place the marker'''
    global position
    player_ch=int(input('Please Enter position for your marker(1-9): '))
    #Value should be between 1-9
    if player_ch>0 and player_ch<10:
        position=player_ch
    return space_check(board, position)    

def space_check(board, position):
    '''Checking if entered position is free or not'''
    if board[position]==' ':
        return True
    else: return False

def place_marker(board, marker, position):
    '''Placing marker at entered position'''
    #Marker will be placed after verifying that position is available
    board[position]=marker
    return board
          
def full_board_check(board):
    '''Checking whether any space is left on the board'''
    #to check if game is tied
    check=False
    for value_check in board:
        if value_check==' ':
            check=True
        else:
            continue
    return check        
        

def display_board(board):
    '''Displaying the board for Game'''
    clear_output()
    print('   |   |   ')
    print(' '+board[7]+' | '+board[8]+' | '+board[9])
    print('   |   |   ')
    print('-----------')
    print('   |   |   ')
    print(' '+board[4]+' | '+board[5]+' | '+board[6])
    print('   |   |   ')
    print('-----------')
    print('   |   |   ')
    print(' '+board[1]+' | '+board[2]+' | '+board[3])
    print('   |   |   ')
    return


def replay():
    '''User choice if they want to play again'''
    replay_choice=input('Do you want to play again? (Y/N): ')
    if replay_choice=='Y':
        return True
    else: return False
    
    
def win_check(board):
    '''Checking if game is won/tie'''
    global game_on
    if ((((board[1]==board[2] and board[1]==board[3]) or (board[1]==board[5] and board[1]==board[9])) or (board[1]==board[4] and board[1]==board[7])) and board[1]!=' '):
        print(board[1]+' has won')
        game_on=False
    elif(board[2]==board[5] and board[2]==board[8] and board[2]!=' '):
        print(board[2]+' has won')
        game_on=False
    elif(board[3]==board[6] and board[3]==board[9] and board[3]!=' '):
        print(board[3]+' has won')
        game_on=False
    elif(board[4]==board[5] and board[4]==board[6] and board[4]!=' '):
        print(board[4]+' has won')
        game_on=False
    elif(((board[7]==board[8] and board[7]==board[9]) or (board[7]==board[5] and board[7]==board[3])) and board[7]!=' '):
        print(board[7]+' has won')
        game_on=False
    elif full_board_check(board)!=True:
        print('Game is Drawn')
        game_on=False
    return game_on


#Main Algorithm including function calls
while play:
    # Set the game up here
    
    #Initialising the Board 
    board = [' ']*10
    
    #Selecting Markers for players
    player1_marker, player2_marker=player_input() 
    
    #Randomly Choose which Player will go first
    Turn=choose_first()
    print(f'{Turn} goes first')
    
    game_on=True
    while game_on:  
        
    #Player 1 Turn
        if Turn=='Player1':       
            print('PLAYER 1:')
            if player_choice(board):
                place_marker(board, player1_marker, position)
                display_board(board)
                Turn='Player2'
            else: print('Position occupied! Enter another position')
    #Player2's Turn
        else:     
            print('PLAYER 2:')
            if player_choice(board):
                place_marker(board, player2_marker, position)
                display_board(board)
                Turn='Player1'
            else: print('Position occupied! Enter another position')
        #check if any player won or tie
        win_check(board)
    if replay():
        play==True
    else:
        break
```
